                         EXPLORATORY DATA ANALYSIS

```python
plt.figure(figsize = (5,4))
ax = sns.countplot(data = df, x='gender')
ax.bar_label(ax.containers[0])
plt.title("Investment Rates By Gender", c ='brown')
plt.show()
```

From dataset male are more likely to invest than female.

```python
plt.figure(figsize=(5,4))
ax = sns.countplot(data = df , x = 'What are your savings objectives?')
ax.bar_label(ax.containers[0])
plt.title('Investment By Savings Objectives', c='brown')
plt.show()
```

From dataset investors are more likely to invest in Retirement Plan than Health Care and Education.

```python
data_counts = df['Source'].value_counts()
plt.pie(data_counts, labels=data_counts.index, autopct='%1.1f%%', startangle=200)
plt.title('sources do investors look at when investing', c='brown')
plt.show()
```

Here we see that investors generally invest through Financial Consultants.

```python
sns.boxplot(data = df , x='age')
plt.title('boxplot')
plt.show()
```

```python
plt.figure(figsize=(6,4))
ax = sns.countplot(data = df , x = 'Duration')
ax.bar_label(ax.containers[0])
plt.title('Number Of Investors By Duration',c='brown')
plt.show()
```

The majority of investors aim for the medium term for 3-5 years duration.

```python
plt.figure(figsize=(5,4))
ax = sns.countplot(data = df , x = 'Invest_Monitor')
ax.bar_label(ax.containers[0])
plt.title('Investment by Invest Monitor',c='brown')
plt.show()
```

The majority of investors prefer to track their investments on a monthly basis.

```python
plt.figure(figsize=(9,6))
sns.pointplot(x='Purpose',y='age',data = df,linestyles='--', capsize=0.5, color='green')
plt.title('pointplot')
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.countplot(x=df['age'],hue=df['gender']) 
plt.title("Investors Gender By Age",c='brown')
plt.show()
```

Men at the ages of 27 and Women at the ages of 24 invest more.

```python
data_counts = df['Avenue'].value_counts()
plt.pie(data_counts, labels=data_counts.index, autopct='%1.1f%%', startangle=200)
plt.title("Distribution of Investment Routes",c='brown')
plt.show()
```

Investors prefer Mutual Funds more than others.

```python
numerical_df = df.select_dtypes(include=[ 'int64'])
sns.heatmap(numerical_df.corr(), annot=True, fmt=".2f")
plt.show()
```

The correlation between Debentures and government bonds is the best correlation fit.

```python
female_expectations = df[df['gender'] == 'Female']['Expect'].value_counts()
male_expectations = df[df['gender'] == 'Male']['Expect'].value_counts()
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))

ax1.pie(female_expectations, labels=female_expectations.index, autopct='%1.1f%%')
ax1.set_title('Expectations of Female Investors',c='brown')

ax2.pie(male_expectations, labels=male_expectations.index, autopct='%1.1f%%')
ax2.set_title('Expectations of Male Investors',c='brown')
plt.show()
```

80% of men and female expect a return of 20%-30% from investments.

```python
investment_columns = ['Mutual_Funds', 'Equity_Market', 'Debentures', 'Government_Bonds', 'Fixed_Deposits', 'PPF', 'Gold']
df[investment_columns].hist(bins=10, figsize=(15, 10), layout=(2, 4))
plt.suptitle('Distribution of Investment Preferences',c='brown')
plt.show()
```

```python
male_df = df[(df['gender'] == 'Male') & (df['Investment_Avenues'] == 'Yes')]
male_investment_age_counts = male_df.groupby('age').size().reset_index(name='count')
plt.figure(figsize=(8, 5))
plt.plot(male_investment_age_counts['age'], male_investment_age_counts['count'], marker='o', linestyle='-')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Number of Male Investors by Age',c='brown')
plt.grid(True)
plt.show()
```

Men are more willing to invest between the ages of 26 and 30.

```python
female_df = df[(df['gender'] == 'Female') & (df['Investment_Avenues'] == 'Yes')]
female_age_counts = female_df.groupby('age').size().reset_index(name='count')
plt.figure(figsize=(10, 6))
plt.plot(female_age_counts['age'], female_age_counts['count'], marker='o', linestyle='-')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Number of Female Investors by Age',c='brown')
plt.grid(True)
plt.show()
```

The age for women to start investing is 28. Women start investing later than men.

```python
plt.figure(figsize=(8, 6))
ax = sns.countplot(data= df,x='Reason_Equity')
plt.title('Reason for Equity Market Investment',c='brown')
ax.bar_label(ax.containers[0])
plt.show()
```

Those who preferred Equity Market said that they chose it because they were investing for Capital Appreciation.

```python
plt.figure(figsize=(8, 6))
ax = sns.countplot(data= df,x='Reason_Mutual')
plt.title('Reason for Mutual Funds Investment',c='brown')
ax.bar_label(ax.containers[0])
plt.show()
```

Those who preferred Mutual Funds said that they chose it because it offered better returns.

```python
plt.figure(figsize=(8, 6))
ax = sns.countplot(data= df,x='Reason_Bonds')
plt.title('Reason for Government Bonds Investment',c='brown')
ax.bar_label(ax.containers[0])
plt.show()
```

Those who preferred Fixed Deposits said that they chose it because it offered Assured Returns.

```python
plt.figure(figsize=(8, 6))
ax = sns.countplot(data= df,x='Reason_FD')
plt.title('Reason for Fixed Deposits Investment',c='brown')
ax.bar_label(ax.containers[0])
plt.show()
```

Those who preferred Fixed Deposits said that they chose it because it offered Fixed Returns.

```python
categories = ['Mutual_Funds', 'Equity_Market', 'Debentures', 'Government_Bonds', 'Fixed_Deposits', 'PPF', 'Gold']
counts = df[categories].sum().reset_index()
counts.columns = ['Category', 'Total']
plt.figure(figsize=(12, 6))
plt.bar(counts['Category'], counts['Total'])
plt.xlabel('Investment Type')
plt.ylabel('Total Investment')
plt.title('Total Investments of Investors in Different Categories', c='brown')
plt.show()
```

Investors preferred Debentures and Gold more than others.

```python
df_financial_choice = df[df["Source"] == "Financial Consultants"]
categories = ['Mutual_Funds', 'Equity_Market', 'Debentures', 'Government_Bonds', 'Fixed_Deposits', 'PPF', 'Gold']
counts = df_financial_choice[categories].sum().reset_index()
counts.columns = ['Category', 'Total']
plt.figure(figsize=(12, 6))
plt.bar(counts['Category'], counts['Total'])
plt.xlabel('Investment Type')
plt.ylabel('Total Investment')
plt.title('Total Investment by Financial Consultants in Different Categories',c='brown')
plt.show()
```

Financial Consultants prefer investing in debentures and gold.
